```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
num_vectors = 100  # Number of vectors
radius = 1  # Radius of the circle
angles = np.linspace(0, 2 * np.pi, num_vectors)

# Compute vector endpoints
x = radius * np.cos(angles)
y = radius * np.sin(angles)

# Define colors for quadrants
colors = []
for i in range(num_vectors):
    if x[i] >= 0 and y[i] >= 0:
        colors.append('blue')  # First quadrant
    elif x[i] < 0 and y[i] >= 0:
        colors.append('red')   # Second quadrant
    elif x[i] < 0 and y[i] < 0:
        colors.append('green') # Third quadrant
    else:
        colors.append('yellow') # Fourth quadrant

# Plot vectors
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-radius - 0.2, radius + 0.2)
ax.set_ylim(-radius - 0.2, radius + 0.2)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.set_aspect('equal')

for i in range(num_vectors):
    ax.quiver(0, 0, x[i], y[i], angles='xy', scale_units='xy', scale=1, color=colors[i])

plt.title("Vectors Forming a Circle with Quadrant-Based Colors")
plt.show()
```

```python
vectors = np.column_stack((x, y))
print(vectors.shape)
```

```python
M = np.array([[1 , 1],
              [1 , 2]])
M
```

```python
new_vectors = np.dot(vectors, M)
new_vectors.shape
```

```python
new_x = new_vectors[:,0]
new_y = new_vectors[:,1]
```

```python
eigenvalues, eigenvectors = np.linalg.eig(M)



fig, ax = plt.subplots(figsize=(8, 8))

ax.set_xlim(-2, 2)
ax.set_ylim(-2.5, 2.5)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.set_aspect('equal')

for i in range(num_vectors):
    ax.quiver(0, 0, new_x[i], new_y[i], angles='xy', scale_units='xy', scale=1,
              color=colors[i])

for i in range(len(eigenvalues)):
    ev = eigenvectors[:, i] * radius
    ax.quiver(0, 0, ev[0], ev[1], angles='xy', scale_units='xy', scale=1,
              color='black', linewidth=2)

plt.title("Vectors after matrix action")
plt.show()
```
